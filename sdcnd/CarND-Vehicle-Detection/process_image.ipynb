```python
from math import ceil
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def plot_images(names, images_per_row, figure_size, cmap='viridis'):

    fig = plt.figure(figsize=figure_size)
    row_count = ceil(len(names) / images_per_row)
    row = 2
    col = 0
    index = 0
    for name in names:
        img = mpimg.imread('output_images/' + name)
        col += 1
        ax = fig.add_subplot(row, len(names), col)
        ax.set_title(name)
        ax.imshow(img, cmap=cmap)
        index += 1
        if col == images_per_row:
            col = 0
            row -= 1
    plt.show()
```

```python
box_images = ['box_image{}.png'.format(i) for i in range(5,11)]
plot_images(box_images, 3, (30,12))
```

```python
heat_images = ['heat_image{}.png'.format(i) for i in range(5,11)]
plot_images(heat_images, 3, (30,12), cmap='hot')
```

```python
label_images = ['label_image10.png']
plot_images(label_images, 3, (30,12), cmap='gray')
```

```python
bbox_images = ['bbox_image10.png']
plot_images(bbox_images, 3, (30,12))
```
